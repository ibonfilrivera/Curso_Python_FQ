## **Ejemplo de operaciones en Química**

¿Cuántas moléculas hay en un gramo de HCl?

- Masa de HCl = 1.0 $g$
- Masa molar de HCl = 36.46 $\frac{g}{mol}$
- Constante de Avogadro = $6.022 \times 10^{23} \frac{1}{mol}$

Resolución:

$ n_{\text{moléculas}} = \text{1.0 g HCl}
\left( \frac{\text{1 mol HCl}}{\text{36.46 g HCl}} \right)
\left( \frac{6.022\times 10^{23} \text{moléculas}}{\text{1 mol HCl}} \right)
= 1.65 \times 10^{22} \text{ moléculas} $

````python
# Definir variables
masa_de_hcl = 1.0 # g
masa_molar_hcl = 36.46 # g/mol
constante_de_avogadro = 6.022 * 10**(23)

# Operaciones
numero_de_moleculas = masa_de_hcl / masa_molar_hcl * constante_de_avogadro

# Imprimir
print(f"Hay {numero_de_moleculas:.0f} moléculas en {masa_de_hcl} g de HCl.")
print(f"Hay {numero_de_moleculas:.4e} moléculas en {masa_de_hcl} g de HCl.")
````

## **El condicional `if`**

Las estructuras `if`, `elif` y `else` se utilizan para controlar el flujo de un programa basándose en **condiciones**:

  - `if`: Ejecuta un bloque de código si una condición es verdadera.
  - `elif` ("else if"): Evalúa una nueva condición si la condición del `if` anterior (o `elif` previo) fue falsa. Pueden haber múltiples `elif`.
  - `else` (opcional): Ejecuta un bloque de código si ninguna de las condiciones `if` o `elif` anteriores fue verdadera.


````python
# Podemos clasificar el estado de agregación de una sustancia

T_fusion = 0  # Temperatura de fusión en °C
T_ebullicion = 100 # Temperatura de ebullición en °C

T = float(input("¿A qué temperatura está el agua?"))

if T > T_ebullicion:
    print("El agua está en estado gaseoso")

elif T < T_fusion:
    print("El agua está en estado sólido")

else:
    print("El agua está en estado líquido")
````

### **Ejercicio 3: Clasificación de enlaces**

Dependiendo de la diferencia de electronegatividad $\left( \Delta \chi \right)$ entre dos átomos, se puede predecir qué tipo de enlace formarán.

- Covalente no polar: $\Delta \chi < 0.5$
- Covalente polar: $0.5 \leq \Delta \chi < 1.7$
- Iónico: $\Delta \chi \geq 1.7$



Escribe una función que reciba la diferencia de electronegatividad entre CO y prediga su tipo de enlace.

````python
electronegatividad_C = 2.5
electronegatividad_O = 3.5

# 1. Restar valor_mayor - valor_menor
diferencia_electronegatividad = electronegatividad_O - electronegatividad_C
print(diferencia_electronegatividad)

# 2. Usar el valor absoluto: abs()
diferencia_electronegatividad = abs(electronegatividad_C - electronegatividad_O)
print(diferencia_electronegatividad)

# Condicionales
if (diferencia_electronegatividad >= 0) and (diferencia_electronegatividad < 0.5):
    print("El enlace es covalente no polar")

elif (diferencia_electronegatividad >= 0.5) and (diferencia_electronegatividad < 1.7):
    print("El enlace es covalente polar")

elif (diferencia_electronegatividad >= 1.7) and (diferencia_electronegatividad < 4.0):
    print("El enlace es iónico")

else:
    print("Diferencia de electronegatividad fuera de rango")
````

## **Práctica dirigida: Ley de los Gases Ideales**

Crearemos nuestra primera función para resolver tareas cotidianas en el laboratorio.

**Objetivo:** Calcular la presión ($P$) despejando la ecuación $PV = nRT$.

$$P = \frac{nRT}{V}$$

donde

- $n$: cantidad de sustancia $(mol)$
- $T$: temperatura $(K)$
- $V$: volumen $(L)$
- $R$: constante universal de los gases $\left( \frac{atm \cdot L}{mol \cdot K} \right)$

A las funciones se les pueden proporcionar distintos argumentos, algunos puedes tener un valor predeterminado (por ejemplo, $R = 0.0821$) y otros se deben proporcionar obligatoriamente ($n$, $T$, $V$).

````python
# Definimos una función que calcula la presión (P)
def calcular_presion(n, R, T, V):
    P = (n*R*T) / V
    #print(f"La presión es {P}")
    return P

# Definimos los argumentos de la función
n = 1.0
T = 298.15 # 25 °C
V = 22.4
R = 0.0821

# Llamar a la función
P = calcular_presion(n, R, T, V)
print(f"La presión es {P:.2f} atm")
````

Podemos asignar diferentes valores a los argumentos al llamar a la función.

````python
# Nuevos argumentos
# calcular_presion(n, R, T, V)

P = calcular_presion(n=2, R=0.0821, T=455, V=1)
print(f"La presión es {P:.2f} atm")
````
