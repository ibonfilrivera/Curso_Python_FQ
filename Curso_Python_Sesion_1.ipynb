## **Ejemplo de operaciones en Química**

¿Cuántas moléculas hay en un gramo de HCl?

- Masa de HCl = 1.0 $g$
- Masa molar de HCl = 36.46 $\frac{g}{mol}$
- Constante de Avogadro = $6.022 \times 10^{23} \frac{1}{mol}$

Resolución:

$ n_{\text{moléculas}} = \text{1.0 g HCl}
\left( \frac{\text{1 mol HCl}}{\text{36.46 g HCl}} \right)
\left( \frac{6.022\times 10^{23} \text{moléculas}}{\text{1 mol HCl}} \right)
= 1.65 \times 10^{22} \text{ moléculas} $

````python
# Definir variables

masa_hcl = 1.0                          # float (g)
masa_molar_hcl = 36.46                  # float (g/mol)
constante_avogadro = 6.022 * 10**(23)   # float (mol⁻¹)
````

````python
# Realizar operaciones

numero_moleculas = masa_hcl / masa_molar_hcl * constante_avogadro
````

````python
# Imprimir resultado

print(f"Hay {numero_moleculas:.2e} moléculas en un gramo de HCl.")
````

## **El condicional `if`**

Las estructuras `if`, `elif` y `else` se utilizan para controlar el flujo de un programa basándose en **condiciones**:

  - `if`: Ejecuta un bloque de código si una condición es verdadera.
  - `elif` ("else if"): Evalúa una nueva condición si la condición del `if` anterior (o `elif` previo) fue falsa. Pueden haber múltiples `elif`.
  - `else`: Ejecuta un bloque de código si ninguna de las condiciones `if` o `elif` anteriores fue verdadera. Es opcional y solo puede haber uno por bloque.


````python
# ¿Cómo podríamos clasificar el estado de agración de una sustancia con base en su temperatura?

Tf = 0    # Temperatura de fusión en °C
Te = 100  # Temperatura de ebullición en °C

T  = int(input('¿A qué temperatura está el agua? (°C): '))   # Introduce el valor

if T > Te:
    print("El agua está en fase gaseosa.")

elif T < Tf: # Solo se ejecuta si la condición de 'if' es falsa
    print("El agua está en fase sólida")

else: # Se ejecuta cuando todas las condiciones anteriores son falsas
    print("El agua está en fase líquida")
````

## **Práctica dirigida: Ley de los Gases Ideales**

Crearemos nuestra primera función para resolver tareas cotidianas en el laboratorio.

**Objetivo:** Calcular la presión ($P$) despejando la ecuación $PV = nRT$.

$$P = \frac{nRT}{V}$$

donde

- $n$: cantidad de sustancia $(mol)$
- $T$: temperatura $(K)$
- $V$: volumen $(L)$
- $R$: constante universal de los gases $\left( \frac{atm \cdot L}{mol \cdot K} \right)$

A las funciones se les pueden proporcionar distintos argumentos, algunos puedes tener un valor predeterminado (por ejemplo, $R = 0.0821$) y otros se deben proporcionar obligatoriamente ($n$, $T$, $V$).

````python
# Definimos una función que calcula la presión de un gas ideal
def calcular_presion(n, T, V, R=0.0821):
    presion = (n * R * T) / V
    return presion # Guarda el cálculo

# Definimos los argumentos que usará la función
n_mol = 1.0
T_K = 298.15 # 25°C
V_L = 22.4

# Llamamos a la función con sus argumentos
P_atm = calcular_presion(n_mol, T_K, V_L) # R ya está predeterminado

# Imprimimos el resultado
print(f"La presión calculada es: {P_atm:.2f} atm")
````

Podemos asignar diferentes valores a los argumentos al llamar a la función.

````python
P = calcular_presion(n=1, T=273.15, V=1)
print(f"La presión calculada es: {P:.2f} atm")
````

````python
# Modificamos el valor de R = 8.3144 (Pa·m³)/(mol·K)

P = calcular_presion(n=1, T=273.15, V=1, R=8.3144)
print(f"La presión calculada es: {P:.2f} Pa")
````

### **Ejercicio 4: Función para calcular temperatura**

Define una función que calcule la temperatura de un gas ideal y determina a qué temperatura se cumplen las siguientes condiciones experimentales para 50.0 g de agua:

- $\text{P = 3.0 atm, V = 20.0 L}$
- $\text{P = 6.3 atm, V = 1.2 L}$

````python
# Escribe tu código

# Calculamos la cantidad de sustancia del agua

m_agua = 10.0
MM_agua = 18.0
n_agua = m_agua / MM_agua

print(f"n = {n_agua:.4f} mol")
````

````python
# Definimos la función que calcula la temperatura

def calcular_temperatura(P, V, n=n_agua, R=0.0821):
    T = (P*V)/(n*R)
    return T
````

````python
# Aplicamos la función cuando P = 3.0 atm, V = 20.0 L

T = calcular_temperatura(P=3.0, V=20.0) # n=n_agua, R=0.0821
print(f"El agua está a {T:.2f} K")
````

````python
# Aplicamos la función cuando P = 6.3 atm, V = 1.2 L

T = calcular_temperatura(P=6.3, V=1.2) # n=n_agua, R=0.0821
print(f"El agua está a {T:.2f} K")
````

**5. Predicción de espontaneidad:** El gas metano reacciona con vapor de agua para producir monóxido de carbono e hidrógeno:

$$CH_4(g) + H_2O(g) \rightarrow CO(g) + H_2(g)$$

Para esta reacción, $\Delta H = 206.1~\frac{kJ}{mol}$ mientras que $\Delta S = 215~\frac{J}{mol\cdot K}$.

Escribe una función que calcule la energía libre de Gibbs de esta reacción a 25 °C con la ecuación $\Delta G = \Delta H - T \Delta S$ y que la clasifique según el caso:

- $\Delta G < 0$: Espontánea
- $\Delta G = 0$: En equilibrio
- $\Delta G > 0$: No espontánea

````python
def calculo_Delta_G(Delta_H, Delta_S, T):
    Delta_G = Delta_H - T * Delta_S

    if Delta_G < 0:
        print("La reacción es espontánea.")
    elif Delta_G == 0:
        print("La reacción está en equilibrio.")
    elif Delta_G > 0:
        print("La reacción no es espontánea.")

    return Delta_G

Delta_H = 206.1
Delta_S = 215 / 1000 # Conversión a kJ/(mol⋅K)
T = 25 + 273.15

Delta_G = calculo_Delta_G(Delta_H, Delta_S, T)

print(f"ΔG = {Delta_G:.2f} kJ/mol")
````
